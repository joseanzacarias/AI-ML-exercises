# Part 1

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_car = pd.read_csv('fuel_efficiency_data.csv')
df_car.head()
```

```python
indp_var = df_car[['Engine_Size_Liters']]
dep_var = df_car['Fuel_Efficiency_MPG']
```

```python
plt.scatter(indp_var, dep_var)
plt.xlabel('Engine Size (L)')
plt.ylabel('Efficiency (MPG)')
plt.title('Ind vs Dep')
plt.show()
```

```python
train_x, test_x, train_y, test_y = train_test_split(indp_var, dep_var, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(train_x, train_y)
```

```python
pred = lr_model.predict(test_x)
```

```python
plt.scatter(train_x, train_y, color='blue', label='Training set')
plt.scatter(test_x, test_y, color='red', label='Test set')
plt.plot(test_x, pred, color='green', label='Pred. Line')
plt.legend()
plt.xlabel('Engine Size (L)')
plt.ylabel('Efficiency (MPG)')
plt.title('Ind vs Dep')
plt.show()
```

```python
new_entry = input("Please enter an engine size in liters: ")
new_entry = np.array([[float(new_entry)]])
pred_new =  lr_model.predict(new_entry)
print("Predicted fuel efficiency (MPG): ", pred_new[0])
```

# Part 2: Linear Regression in Python

#### Importing dataset

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

df_robotic = pd.read_csv('Robotic Data.csv')
df_robotic.head()
```

```python
robot_indp_var = df_robotic[['Factor X']]
robot_dep_var = df_robotic['Robot efficiency index']
plt.scatter(robot_indp_var, robot_dep_var)
plt.xlabel('Factor X')
plt.ylabel('Efficiency index')
plt.title('Factor X vs Efficiency Index')
plt.show()
```

```python
robot_train_x, robot_test_x, robot_train_y, robot_test_y = train_test_split(robot_indp_var, robot_dep_var, test_size=0.3, random_state=30)
robot_lr_model = LinearRegression()
robot_lr_model.fit(robot_train_x, robot_train_y)
robot_pred = robot_lr_model.predict(robot_test_x)
plt.scatter(robot_train_x, robot_train_y, color='blue', label='Training set')
plt.scatter(robot_test_x, robot_test_y, color='red', label='Test set')
plt.plot(robot_test_x, robot_pred, color='green', label='Pred. Line')
plt.legend()
plt.xlabel('Factor X')
plt.ylabel('Efficiency index')
plt.title('Factor X vs Efficiency Index')
plt.show()
```
